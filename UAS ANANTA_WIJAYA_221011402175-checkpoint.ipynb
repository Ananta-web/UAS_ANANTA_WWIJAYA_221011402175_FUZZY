```python
#1.SISTEM PENENTUA TOKO ONLINE SAYA MENGGUNAKAN SISTEMNYA KODE PROMO?
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# 1. Definisi Variabel Input (Jumlah Pembelian dan Kode Promo)
jumlah_pembelian = ctrl.Antecedent(np.arange(0, 10001, 1), 'jumlah_pembelian')
kode_promo = ctrl.Antecedent(np.arange(0, 2, 1), 'kode_promo')  # 0 = Tidak Valid, 1 = Valid

# 2. Definisi Variabel Output (Diskon)
diskon = ctrl.Consequent(np.arange(0, 51, 1), 'diskon')

# 3. Definisi Himpunan Fuzzy untuk Jumlah Pembelian
jumlah_pembelian['sedikit'] = fuzz.trapmf(jumlah_pembelian.universe, [0, 0, 1000, 2000])
jumlah_pembelian['sedang'] = fuzz.trapmf(jumlah_pembelian.universe, [1000, 2000, 4000, 5000])
jumlah_pembelian['banyak'] = fuzz.trapmf(jumlah_pembelian.universe, [4000, 5000, 10000, 10000])

# 4. Definisi Himpunan Fuzzy untuk Kode Promo
kode_promo['tidak_valid'] = fuzz.trimf(kode_promo.universe, [0, 0, 0])
kode_promo['valid'] = fuzz.trimf(kode_promo.universe, [0, 1, 1])

# 5. Definisi Himpunan Fuzzy untuk Diskon
diskon['kecil'] = fuzz.trapmf(diskon.universe, [0, 0, 10, 15])
diskon['sedang'] = fuzz.trapmf(diskon.universe, [10, 15, 25, 30])
diskon['besar'] = fuzz.trapmf(diskon.universe, [25, 30, 50, 50])

# 6. Aturan Fuzzy (Rule Base)
rule1 = ctrl.Rule(jumlah_pembelian['sedikit'] & kode_promo['tidak_valid'], diskon['kecil'])
rule2 = ctrl.Rule(jumlah_pembelian['sedang'] & kode_promo['tidak_valid'], diskon['sedang'])
rule3 = ctrl.Rule(jumlah_pembelian['banyak'] & kode_promo['tidak_valid'], diskon['besar'])

rule4 = ctrl.Rule(jumlah_pembelian['sedikit'] & kode_promo['valid'], diskon['sedang'])
rule5 = ctrl.Rule(jumlah_pembelian['sedang'] & kode_promo['valid'], diskon['besar'])
rule6 = ctrl.Rule(jumlah_pembelian['banyak'] & kode_promo['valid'], diskon['besar'])

# 7. Sistem Kontrol Fuzzy
diskon_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6])
diskon_simulasi = ctrl.ControlSystemSimulation(diskon_ctrl)

# 8. Fungsi untuk Menghitung Diskon Berdasarkan Jumlah Pembelian dan Kode Promo
def hitung_diskon(jumlah, promo):
    # Jika promo valid, inputkan 1, jika tidak valid, inputkan 0
    promo_input = 1 if promo.lower() == 'valid' else 0
    diskon_simulasi.input['jumlah_pembelian'] = jumlah
    diskon_simulasi.input['kode_promo'] = promo_input
    diskon_simulasi.compute()
    return diskon_simulasi.output['diskon']

# 9. Contoh Simulasi
jumlah = 7000  # Masukkan jumlah pembelian
promo = 'valid'  # Masukkan kode promo, 'valid' atau 'tidak valid'

hasil_diskon = hitung_diskon(jumlah, promo)
print(f"Jumlah Pembelian: {jumlah}")
print(f"Kode Promo: {promo}")
print(f"Diskon yang Diterima: {hasil_diskon:.2f}%")

# 10. Visualisasi Himpunan Fuzzy
jumlah_pembelian.view()
kode_promo.view()
diskon.view()

# 11. Menampilkan Visualisasi pada Plot
plt.show()
```

```python
#2.SISTEM MENENTUKAN TINGKAT DISKON BERDASARKAN PARAMETER JUMLAH PEMBELIAN DAN FREKUENSI PELANGGAN?
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# 1. Definisi Variabel Input (Jumlah Pembelian dan Frekuensi Belanja)
jumlah_pembelian = ctrl.Antecedent(np.arange(0, 10001, 1), 'jumlah_pembelian')
frekuensi_belanja = ctrl.Antecedent(np.arange(0, 31, 1), 'frekuensi_belanja')

# 2. Definisi Variabel Output (Diskon)
diskon = ctrl.Consequent(np.arange(0, 51, 1), 'diskon')

# 3. Definisi Himpunan Fuzzy untuk Jumlah Pembelian
jumlah_pembelian['sedikit'] = fuzz.trapmf(jumlah_pembelian.universe, [0, 0, 1000, 2000])
jumlah_pembelian['sedang'] = fuzz.trapmf(jumlah_pembelian.universe, [1000, 2000, 4000, 5000])
jumlah_pembelian['banyak'] = fuzz.trapmf(jumlah_pembelian.universe, [4000, 5000, 10000, 10000])

# 4. Definisi Himpunan Fuzzy untuk Frekuensi Belanja
frekuensi_belanja['jarang'] = fuzz.trapmf(frekuensi_belanja.universe, [0, 0, 5, 10])
frekuensi_belanja['rata-rata'] = fuzz.trapmf(frekuensi_belanja.universe, [5, 10, 15, 20])
frekuensi_belanja['sering'] = fuzz.trapmf(frekuensi_belanja.universe, [15, 20, 30, 30])

# 5. Definisi Himpunan Fuzzy untuk Diskon
diskon['kecil'] = fuzz.trapmf(diskon.universe, [0, 0, 10, 15])
diskon['sedang'] = fuzz.trapmf(diskon.universe, [10, 15, 25, 30])
diskon['besar'] = fuzz.trapmf(diskon.universe, [25, 30, 50, 50])

# 6. Aturan Fuzzy (Rule Base)
rule1 = ctrl.Rule(jumlah_pembelian['sedikit'] & frekuensi_belanja['jarang'], diskon['kecil'])
rule2 = ctrl.Rule(jumlah_pembelian['sedang'] & frekuensi_belanja['rata-rata'], diskon['sedang'])
rule3 = ctrl.Rule(jumlah_pembelian['banyak'] & frekuensi_belanja['sering'], diskon['besar'])
rule4 = ctrl.Rule(jumlah_pembelian['banyak'] & frekuensi_belanja['rata-rata'], diskon['sedang'])
rule5 = ctrl.Rule(jumlah_pembelian['sedang'] & frekuensi_belanja['sering'], diskon['besar'])

# 7. Sistem Kontrol Fuzzy
diskon_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
diskon_simulasi = ctrl.ControlSystemSimulation(diskon_ctrl)

# 8. Fungsi untuk Menghitung Diskon Berdasarkan Input
def hitung_diskon(jumlah, frekuensi):
    diskon_simulasi.input['jumlah_pembelian'] = jumlah
    diskon_simulasi.input['frekuensi_belanja'] = frekuensi
    diskon_simulasi.compute()
    return diskon_simulasi.output['diskon']

# 9. Contoh Simulasi
jumlah = 7000  # Masukkan jumlah pembelian
frekuensi = 18  # Masukkan frekuensi belanja

hasil_diskon = hitung_diskon(jumlah, frekuensi)
print(f"Jumlah Pembelian: {jumlah}")
print(f"Frekuensi Belanja: {frekuensi}")
print(f"Diskon yang Diterima: {hasil_diskon:.2f}%")

# 10. Visualisasi Himpunan Fuzzy
jumlah_pembelian.view()
frekuensi_belanja.view()
diskon.view()

# 11. Menampilkan Visualisasi pada Plot
plt.show()
```
